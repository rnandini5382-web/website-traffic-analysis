# 📚 Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

# 📂 Loading the Dataset

```python
df = pd.read_csv('/content/traffic.csv')
```

```python
df.head()
```

# 🔍 Understanding the Dataset

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

# 🧹 Data Cleaning & Preprocessing

```python
df.drop_duplicates(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

# 📊 KPI Analysis

## **TOTAL EVENTS**

```python
total_events = df['event'].count()

print("Total Events:", total_events)
```

## **TOTAL COUNTRIES**

```python
total_countries = df['country'].nunique()

print("Total Countries:", total_countries)
```

### **TOTAL CITIES**

```python
total_cities = df['city'].nunique()

print("Total Cities:", total_cities)
```

## **TOTAL ARTISTS**

```python
total_artists = df['artist'].nunique()

print("Total Artists:", total_artists)
```

### **Insight**
The KPI analysis provides an overview of total traffic events, geographic reach, artist engagement, and audience interaction patterns within the dataset.

# 🌍 Top Countries Analysis

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Events')
plt.show()
```

### **Business Insight**
Certain countries generate significantly higher traffic activity, indicating stronger audience engagement and platform reach in those regions.

# 🏙️ Top Cities Analysis

```python
top_cities = df['city'].value_counts().head(10)

top_cities.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Number of Events')
plt.show()
```

### **Business Insight**
The analysis identifies cities with the highest user interaction and traffic activity, helping understand audience concentration.

# 🎵 Top Artists Analysis

```python
top_artists = df['artist'].value_counts().head(10)

top_artists.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Traffic Count')
plt.show()
```

### **Business Insight**
Popular artists attract higher user traffic and engagement, indicating strong audience interest and content popularity.

# 🎧 Top Tracks Analysis

```python
top_tracks = df['track'].value_counts().head(10)

top_tracks.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Tracks')
plt.xlabel('Track')
plt.ylabel('Traffic Count')
plt.show()
```

### **Business Insight**
Certain tracks generate significantly higher engagement and user interaction, reflecting trending content and listener preferences.

# 📈 Event Distribution Analysis

```python
event_distribution = df['event'].value_counts()

event_distribution.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))
plt.title('Event Distribution')
plt.ylabel('')
plt.show()
```

### **Business Insight**
Event distribution analysis helps understand user activity behavior and interaction patterns on the platform.

# 📅 Daily Traffic Trend Analysis

```python
daily_traffic = df.groupby('date')['event'].count()

plt.figure(figsize=(12,5))
daily_traffic.plot()
plt.title('Daily Traffic Trend')
plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.show()
```

###**Business Insight**
Traffic activity fluctuates over time, indicating changing user engagement patterns and audience interaction trends.

# 🔥 Missing Values Heatmap

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```

### **Business Insight**
The heatmap helps visualize missing data patterns and improves understanding of dataset quality.
